## 1. Why Type Hints?

Type hints provide:
- **Documentation**: Code is self-documenting
- **IDE Support**: Better autocomplete and refactoring
- **Error Detection**: Catch bugs before runtime with mypy/pyright
- **Readability**: Easier to understand function signatures

````python
# WITHOUT type hints - unclear what types are expected
def process_data(data, threshold, callback):
    results = []
    for item in data:
        if item > threshold:
            results.append(callback(item))
    return results

# What is data? A list of what?
# What is threshold? int? float?
# What is callback? What does it return?
````

````python
from typing import Callable

# WITH type hints - crystal clear!
def process_data(
    data: list[float],
    threshold: float,
    callback: Callable[[float], str]
) -> list[str]:
    """Process numeric data above threshold."""
    results: list[str] = []
    for item in data:
        if item > threshold:
            results.append(callback(item))
    return results

# Now we know:
# - data is a list of floats
# - threshold is a float
# - callback takes a float and returns a string
# - function returns a list of strings
````

### Important: Type Hints Don't Enforce at Runtime!

Python does NOT check types at runtime. Type hints are for:
- Static analysis tools (mypy, pyright)
- IDE support
- Documentation

````python
def greet(name: str) -> str:
    return f"Hello, {name}!"

# This runs without error (but mypy would catch it)
result = greet(123)  # Passing int instead of str
print(result)  # Hello, 123!
````

````python
# =============================================================================
# TYPE ALIAS EXAMPLES
# =============================================================================

from typing import Callable

# Simple aliases
UserId = int
Username = str
Email = str

# Collection aliases
UserIds = list[UserId]
UserMap = dict[UserId, Username]
Coordinates = tuple[float, float]
CoordinateList = list[Coordinates]

# Function aliases
Validator = Callable[[str], bool]
Transformer = Callable[[str], str]
FilterFunc = Callable[[list], list]
Callback = Callable[[], None]

# Complex nested aliases
JsonValue = str | int | float | bool | None | list["JsonValue"] | dict[str, "JsonValue"]
ApiResponse = dict[str, JsonValue]
````

## 4. Callable - Typing Functions

`Callable` lets you type-hint variables that hold functions.

````python
# =============================================================================
# CALLABLE SYNTAX
# =============================================================================

from typing import Callable

# Callable[[INPUT_TYPES], RETURN_TYPE]
#          ↑              ↑
#          Parameters     Return type
#          (in a list)

# Examples:
NoParams: Callable[[], None]                      # def f() -> None
OneParam: Callable[[str], int]                    # def f(s: str) -> int
TwoParams: Callable[[int, int], float]            # def f(a: int, b: int) -> float
ThreeParams: Callable[[str, int, bool], str]      # def f(s: str, n: int, b: bool) -> str
ListToList: Callable[[list], list]                # def f(lst: list) -> list
````

````python
# =============================================================================
# BASIC CALLABLE USAGE
# =============================================================================

from typing import Callable

def add(a: int, b: int) -> int:
    return a + b

def multiply(a: int, b: int) -> int:
    return a * b

# Type alias for math operations
MathOperation = Callable[[int, int], int]

# Variable that holds a function
operation: MathOperation = add
print(operation(5, 3))  # 8

operation = multiply
print(operation(5, 3))  # 15

# List of functions
operations: list[MathOperation] = [add, multiply]
for op in operations:
    print(op(10, 2))  # 12, 20
````

````python
# =============================================================================
# CALLABLE AS FUNCTION PARAMETER
# =============================================================================

from typing import Callable

def apply_operation(
    a: int,
    b: int,
    operation: Callable[[int, int], int]  # Function parameter
) -> int:
    """Apply any operation that takes two ints."""
    return operation(a, b)


def add(x: int, y: int) -> int:
    return x + y

def subtract(x: int, y: int) -> int:
    return x - y


# Pass different functions
result1 = apply_operation(10, 5, add)       # 15
result2 = apply_operation(10, 5, subtract)  # 5

# Pass lambda
result3 = apply_operation(10, 5, lambda a, b: a * b)  # 50

print(f"Add: {result1}, Subtract: {result2}, Multiply: {result3}")
````

````python
# =============================================================================
# CALLABLE AS RETURN TYPE
# =============================================================================

from typing import Callable

def create_multiplier(factor: int) -> Callable[[int], int]:
    """Return a function that multiplies by factor."""
    def multiplier(x: int) -> int:
        return x * factor
    return multiplier


# Create specialized functions
double: Callable[[int], int] = create_multiplier(2)
triple: Callable[[int], int] = create_multiplier(3)

print(double(5))  # 10
print(triple(5))  # 15
````

````python
# =============================================================================
# CALLABLE IN DICTIONARIES (DISPATCH PATTERN)
# =============================================================================

from typing import Callable

# Type alias
FilterFunc = Callable[[list], list]

def grayscale(pixels: list) -> list:
    print("Applying grayscale...")
    return pixels

def sepia(pixels: list) -> list:
    print("Applying sepia...")
    return pixels

def blur(pixels: list) -> list:
    print("Applying blur...")
    return pixels

# Typed dictionary of functions
FILTERS: dict[str, FilterFunc] = {
    "grayscale": grayscale,
    "sepia": sepia,
    "blur": blur,
}

def apply_filter(name: str, pixels: list) -> list:
    """Apply filter by name."""
    if name not in FILTERS:
        raise ValueError(f"Unknown filter: {name}")
    
    filter_func: FilterFunc = FILTERS[name]
    return filter_func(pixels)


# Usage
apply_filter("grayscale", [])  # Applying grayscale...
````

````python
# =============================================================================
# CALLABLE WITH OPTIONAL/DEFAULT PARAMETERS
# =============================================================================

from typing import Callable

# Callable doesn't capture optional params well
# For complex signatures, use Protocol (see Section 9)

def greet(name: str, greeting: str = "Hello") -> str:
    return f"{greeting}, {name}!"

# This works but doesn't show the default parameter
Greeter = Callable[[str, str], str]

# Better: just use ... for "any arguments"
AnyFunc = Callable[..., str]  # Any params, returns str

func: AnyFunc = greet
print(func("Manuel"))            # Hello, Manuel!
print(func("Manuel", "Hi"))      # Hi, Manuel!
````

````python
# =============================================================================
# CALLABLE IN CLASSES
# =============================================================================

from typing import Callable

class EventEmitter:
    """Class that uses Callable for event handlers."""
    
    # Type alias for event handlers
    EventHandler = Callable[[str, dict], None]
    
    def __init__(self) -> None:
        # Dict of event name -> list of handlers
        self._handlers: dict[str, list[EventEmitter.EventHandler]] = {}
    
    def on(self, event: str, handler: EventHandler) -> None:
        """Register an event handler."""
        if event not in self._handlers:
            self._handlers[event] = []
        self._handlers[event].append(handler)
    
    def emit(self, event: str, data: dict) -> None:
        """Emit an event to all handlers."""
        if event in self._handlers:
            for handler in self._handlers[event]:
                handler(event, data)


# Usage
def log_event(event: str, data: dict) -> None:
    print(f"Event: {event}, Data: {data}")

emitter = EventEmitter()
emitter.on("user_login", log_event)
emitter.emit("user_login", {"user": "Manuel"})  # Event: user_login, Data: {'user': 'Manuel'}
````

## 5. TypedDict - Typed Dictionaries

`TypedDict` defines the structure of dictionaries with specific keys and value types.

````python
# =============================================================================
# WHY TYPEDDICT?
# =============================================================================

# Regular dict - no key/value structure enforced
user: dict[str, str | int] = {
    "name": "Manuel",
    "age": 30,
    "typo_key": "oops"  # No error - any key allowed
}

# What keys are expected?
# What types should each key have?
# No way to know without reading the code!
````

````python
# =============================================================================
# BASIC TYPEDDICT
# =============================================================================

from typing import TypedDict

class User(TypedDict):
    """User data structure."""
    name: str
    age: int
    email: str


# ✅ Correct usage
user: User = {
    "name": "Manuel",
    "age": 30,
    "email": "manuel@example.com"
}

# ❌ Type checker catches these errors:
# user: User = {"name": "Manuel"}  # Missing keys
# user: User = {"name": "Manuel", "age": "thirty", "email": "x"}  # Wrong type
# user: User = {"name": "Manuel", "age": 30, "email": "x", "extra": 1}  # Extra key

print(user["name"])  # Manuel
print(type(user))    # <class 'dict'> - Still a regular dict at runtime!
````

````python
# =============================================================================
# OPTIONAL KEYS WITH total=False
# =============================================================================

from typing import TypedDict

# All keys optional
class Config(TypedDict, total=False):
    """All keys are optional."""
    debug: bool
    timeout: int
    retries: int


# All valid:
config1: Config = {}                              # Empty
config2: Config = {"debug": True}                 # Partial
config3: Config = {"debug": True, "timeout": 30}  # More keys
config4: Config = {"debug": True, "timeout": 30, "retries": 3}  # All keys

print(config2)
````

````python
# =============================================================================
# MIXING REQUIRED AND OPTIONAL KEYS (Inheritance)
# =============================================================================

from typing import TypedDict

# Python 3.9-3.10: Use inheritance
class _UserRequired(TypedDict):
    """Required fields."""
    id: int
    name: str

class User(_UserRequired, total=False):
    """User with required and optional fields."""
    email: str
    phone: str
    age: int


# ✅ Valid - required keys present
user1: User = {"id": 1, "name": "Manuel"}
user2: User = {"id": 2, "name": "Alice", "email": "alice@example.com"}

# ❌ Invalid - missing required key
# user3: User = {"name": "Bob"}  # Missing 'id'

print(user1)
print(user2)
````

````python
# =============================================================================
# PYTHON 3.11+ Required AND NotRequired
# =============================================================================

from typing import TypedDict, Required, NotRequired

class User(TypedDict, total=False):
    """User with explicit required/optional markers."""
    id: Required[int]           # Must be present
    name: Required[str]         # Must be present
    email: NotRequired[str]     # Optional
    phone: NotRequired[str]     # Optional


# Same as inheritance approach, but clearer
user: User = {"id": 1, "name": "Manuel"}
print(user)
````

````python
# =============================================================================
# NESTED TYPEDDICT
# =============================================================================

from typing import TypedDict

class Address(TypedDict):
    """Address structure."""
    street: str
    city: str
    country: str
    zip_code: str

class Company(TypedDict):
    """Company structure."""
    name: str
    industry: str

class Employee(TypedDict):
    """Employee with nested structures."""
    id: int
    name: str
    address: Address       # Nested TypedDict
    company: Company       # Nested TypedDict
    skills: list[str]      # Regular list


employee: Employee = {
    "id": 1,
    "name": "Manuel",
    "address": {
        "street": "123 Main St",
        "city": "Austin",
        "country": "USA",
        "zip_code": "78701"
    },
    "company": {
        "name": "Tech Corp",
        "industry": "Software"
    },
    "skills": ["Python", "SQL", "Data Analysis"]
}

print(employee["address"]["city"])  # Austin
````

````python
# =============================================================================
# TYPEDDICT VS DATACLASS
# =============================================================================

from typing import TypedDict
from dataclasses import dataclass

# TypedDict - dict at runtime
class UserDict(TypedDict):
    name: str
    age: int

# Dataclass - custom class at runtime
@dataclass
class UserClass:
    name: str
    age: int


# Usage comparison
user_dict: UserDict = {"name": "Manuel", "age": 30}
user_class = UserClass(name="Manuel", age=30)

# Access
print(user_dict["name"])   # Dict access with []
print(user_class.name)      # Attribute access with .

# Type at runtime
print(type(user_dict))      # <class 'dict'>
print(type(user_class))     # <class '__main__.UserClass'>

# JSON serialization
import json
print(json.dumps(user_dict))  # Works directly!
# print(json.dumps(user_class))  # TypeError - needs custom encoder
````

### When to Use TypedDict vs Dataclass

| Feature | TypedDict | Dataclass |
|---------|-----------|----------|
| Runtime type | `dict` | Custom class |
| Access style | `d["key"]` | `obj.attr` |
| JSON serialization | Works directly | Needs encoder |
| Methods | No | Yes |
| Validation | No | No (use Pydantic) |
| Best for | API responses, config | Domain objects |

## 6. Final - Immutable Values

`Final` indicates a value should not be reassigned.

````python
# =============================================================================
# FINAL FOR CONSTANTS
# =============================================================================

from typing import Final

# Module-level constants
MAX_CONNECTIONS: Final[int] = 100
API_VERSION: Final[str] = "v2.0"
DEBUG_MODE: Final[bool] = False
ALLOWED_EXTENSIONS: Final[tuple[str, ...]] = (".jpg", ".png", ".bmp")

# Type checker warns on reassignment:
# MAX_CONNECTIONS = 200  # Error: Cannot assign to final name

print(f"Max connections: {MAX_CONNECTIONS}")
````

````python
# =============================================================================
# FINAL IN CLASSES
# =============================================================================

from typing import Final

class Config:
    """Application configuration."""
    
    # Class-level constants
    VERSION: Final[str] = "1.0.0"
    MAX_RETRIES: Final[int] = 3
    
    def __init__(self, name: str) -> None:
        # Instance-level final (assigned once)
        self.name: Final[str] = name
    
    def show(self) -> None:
        print(f"{self.name} v{self.VERSION}")


config = Config("MyApp")
config.show()  # MyApp v1.0.0

# Type checker warns:
# config.name = "Other"  # Cannot assign to final attribute
# Config.VERSION = "2.0"  # Cannot assign to final name
````

````python
# =============================================================================
# FINAL DOES NOT PREVENT MUTATION!
# =============================================================================

from typing import Final

# Final prevents REASSIGNMENT, not MUTATION
ITEMS: Final[list[str]] = ["a", "b", "c"]

# ❌ Cannot reassign
# ITEMS = ["x", "y"]  # Type error

# ✅ CAN mutate (Final doesn't prevent this!)
ITEMS.append("d")  # Works!
ITEMS[0] = "z"     # Works!

print(ITEMS)  # ['z', 'b', 'c', 'd']

# For true immutability, use tuple instead of list:
IMMUTABLE_ITEMS: Final[tuple[str, ...]] = ("a", "b", "c")
# IMMUTABLE_ITEMS[0] = "z"  # TypeError at runtime!
````

````python
# =============================================================================
# FINAL WITH TYPE INFERENCE
# =============================================================================

from typing import Final

# Type can be inferred
NAME: Final = "Manuel"       # Inferred as Final[str]
COUNT: Final = 42            # Inferred as Final[int]
PI: Final = 3.14159          # Inferred as Final[float]

# Explicit is better for documentation
NAME_EXPLICIT: Final[str] = "Manuel"
COUNT_EXPLICIT: Final[int] = 42
````

## 7. Union and Optional

For values that can be one of several types.

````python
# =============================================================================
# UNION - Multiple Possible Types
# =============================================================================

from typing import Union

# Old syntax (Python 3.9 and earlier)
value: Union[int, str] = 42
value = "hello"  # Also valid

# New syntax (Python 3.10+) - preferred!
value: int | str = 42
value = "hello"  # Also valid

# Multiple types
number: int | float | complex = 3.14

# In function parameters
def double(value: int | float) -> int | float:
    return value * 2

print(double(5))    # 10
print(double(3.5))  # 7.0
````

````python
# =============================================================================
# OPTIONAL - Value or None
# =============================================================================

from typing import Optional

# Old syntax
name: Optional[str] = None  # Same as Union[str, None]

# New syntax (Python 3.10+) - preferred!
name: str | None = None

# Common use: optional parameters
def greet(name: str | None = None) -> str:
    if name is None:
        return "Hello, stranger!"
    return f"Hello, {name}!"

print(greet())          # Hello, stranger!
print(greet("Manuel"))  # Hello, Manuel!
````

````python
# =============================================================================
# TYPE NARROWING WITH Union
# =============================================================================

def process(value: int | str | list) -> str:
    """Process different types differently."""
    
    # isinstance narrows the type
    if isinstance(value, int):
        # Type checker knows: value is int here
        return f"Integer: {value * 2}"
    
    elif isinstance(value, str):
        # Type checker knows: value is str here
        return f"String: {value.upper()}"
    
    else:
        # Type checker knows: value is list here
        return f"List with {len(value)} items"


print(process(42))           # Integer: 84
print(process("hello"))      # String: HELLO
print(process([1, 2, 3]))    # List with 3 items
````

## 8. Generic Types

Create reusable type-safe containers and functions.

````python
# =============================================================================
# GENERIC FUNCTIONS
# =============================================================================

from typing import TypeVar

# Define a type variable
T = TypeVar('T')  # Can be any type

def first(items: list[T]) -> T:
    """Return first item of any list type."""
    return items[0]


# Type checker infers return type from input
num: int = first([1, 2, 3])        # T = int
text: str = first(["a", "b"])     # T = str
point: tuple = first([(1, 2)])     # T = tuple

print(num, text, point)  # 1 a (1, 2)
````

````python
# =============================================================================
# CONSTRAINED TYPE VARIABLES
# =============================================================================

from typing import TypeVar

# T can only be int or float
Number = TypeVar('Number', int, float)

def add(a: Number, b: Number) -> Number:
    return a + b


result1 = add(1, 2)      # OK: int + int
result2 = add(1.5, 2.5)  # OK: float + float
# result3 = add("a", "b")  # Error: str not allowed

print(result1, result2)  # 3 4.0
````

````python
# =============================================================================
# GENERIC CLASSES
# =============================================================================

from typing import TypeVar, Generic

T = TypeVar('T')

class Stack(Generic[T]):
    """A generic stack that works with any type."""
    
    def __init__(self) -> None:
        self._items: list[T] = []
    
    def push(self, item: T) -> None:
        self._items.append(item)
    
    def pop(self) -> T:
        return self._items.pop()
    
    def peek(self) -> T:
        return self._items[-1]
    
    def is_empty(self) -> bool:
        return len(self._items) == 0


# Create typed stacks
int_stack: Stack[int] = Stack()
int_stack.push(1)
int_stack.push(2)
print(int_stack.pop())  # 2 (type: int)

str_stack: Stack[str] = Stack()
str_stack.push("hello")
str_stack.push("world")
print(str_stack.pop())  # world (type: str)
````

````python
# =============================================================================
# MULTIPLE TYPE VARIABLES
# =============================================================================

from typing import TypeVar, Generic

K = TypeVar('K')  # Key type
V = TypeVar('V')  # Value type

class Pair(Generic[K, V]):
    """A pair of two values with different types."""
    
    def __init__(self, key: K, value: V) -> None:
        self.key = key
        self.value = value
    
    def swap(self) -> "Pair[V, K]":
        return Pair(self.value, self.key)


# Different type combinations
pair1: Pair[str, int] = Pair("age", 30)
pair2: Pair[int, list[str]] = Pair(1, ["a", "b"])

print(f"{pair1.key}: {pair1.value}")  # age: 30

swapped = pair1.swap()  # Type: Pair[int, str]
print(f"{swapped.key}: {swapped.value}")  # 30: age
````

## 9. Protocol - Structural Typing

Protocol defines interfaces based on structure, not inheritance.

````python
# =============================================================================
# PROTOCOL - DEFINE INTERFACE BY STRUCTURE
# =============================================================================

from typing import Protocol

# Define what methods/attributes a type must have
class Drawable(Protocol):
    """Any type that has a draw method."""
    
    def draw(self) -> str:
        """Draw and return representation."""
        ...


# These classes DON'T inherit from Drawable,
# but they match its structure!

class Circle:
    def __init__(self, radius: float) -> None:
        self.radius = radius
    
    def draw(self) -> str:
        return f"Circle with radius {self.radius}"

class Rectangle:
    def __init__(self, width: float, height: float) -> None:
        self.width = width
        self.height = height
    
    def draw(self) -> str:
        return f"Rectangle {self.width}x{self.height}"


# Function accepts anything with draw() method
def render(shape: Drawable) -> None:
    print(shape.draw())


# Both work even though they don't inherit from Drawable!
render(Circle(5.0))        # Circle with radius 5.0
render(Rectangle(3, 4))    # Rectangle 3x4
````

````python
# =============================================================================
# PROTOCOL FOR CALLABLE WITH SPECIFIC SIGNATURE
# =============================================================================

from typing import Protocol

# Better than Callable for complex signatures
class FilterFunction(Protocol):
    """Protocol for image filter functions."""
    
    def __call__(
        self,
        pixels: list[list[list[int]]],
        intensity: float = 1.0
    ) -> list[list[list[int]]]:
        """Apply filter with optional intensity."""
        ...


def grayscale(
    pixels: list[list[list[int]]],
    intensity: float = 1.0
) -> list[list[list[int]]]:
    """Convert to grayscale."""
    print(f"Grayscale at {intensity} intensity")
    return pixels


def apply_filter(filter_func: FilterFunction, pixels: list) -> list:
    return filter_func(pixels, intensity=0.5)


apply_filter(grayscale, [])  # Grayscale at 0.5 intensity
````

## 10. Literal Types

Restrict values to specific literals.

````python
# =============================================================================
# LITERAL - SPECIFIC VALUES ONLY
# =============================================================================

from typing import Literal

# Only these exact values are allowed
Direction = Literal["north", "south", "east", "west"]
StatusCode = Literal[200, 404, 500]
OnOff = Literal[True, False]  # Same as bool, but explicit

def move(direction: Direction) -> None:
    print(f"Moving {direction}")

move("north")  # OK
move("south")  # OK
# move("up")   # Type error: "up" not in Direction
````

````python
# =============================================================================
# LITERAL IN FUNCTION OVERLOADS
# =============================================================================

from typing import Literal, overload

# Use overload when return Type depends on input Type for type checker!

@overload
def process(mode: Literal["text"]) -> str: ...

@overload
def process(mode: Literal["binary"]) -> bytes: ...

def process(mode: Literal["text", "binary"]) -> str | bytes:
    if mode == "text":
        return "text data"
    else:
        return b"binary data"


# Type checker knows exact return type!
text_result = process("text")      # Type: str
binary_result = process("binary")  # Type: bytes

print(text_result, binary_result)
````

## 11. Type Guards

Custom functions that narrow types.

````python
# =============================================================================
# TYPEGUARD - CUSTOM TYPE NARROWING
# =============================================================================

from typing import TypeGuard

def is_string_list(val: list) -> TypeGuard[list[str]]:
    """Check if all items in list are strings."""
    return all(isinstance(item, str) for item in val)


def process(items: list) -> None:
    if is_string_list(items):
        # Type checker knows: items is list[str] here!
        for item in items:
            print(item.upper())  # .upper() is valid
    else:
        print("Not a string list")


process(["hello", "world"])  # HELLO WORLD
process([1, 2, 3])            # Not a string list
````

## 12. Best Practices

````python
# =============================================================================
# TYPE HINTS BEST PRACTICES
# =============================================================================

from typing import TypeAlias, Final, Callable
from pathlib import Path

# 1. USE TYPE ALIASES FOR COMPLEX TYPES
# ❌ Bad
def process_bad(data: list[dict[str, list[tuple[int, int]]]]) -> list[dict[str, list[tuple[int, int]]]]:
    pass

# ✅ Good
Coordinate: TypeAlias = tuple[int, int]
CoordinateList: TypeAlias = list[Coordinate]
ShapeData: TypeAlias = dict[str, CoordinateList]

def process_good(data: list[ShapeData]) -> list[ShapeData]:
    pass


# 2. USE | INSTEAD OF Union (Python 3.10+)
# ❌ Old
from typing import Union, Optional
value_old: Union[int, str] = 42
name_old: Optional[str] = None

# ✅ New
value_new: int | str = 42
name_new: str | None = None


# 3. USE FINAL FOR CONSTANTS
MAX_SIZE: Final[int] = 1000
API_URL: Final[str] = "https://api.example.com"


# 4. PREFER SPECIFIC TYPES OVER Any
# ❌ Bad
from typing import Any
def bad_func(data: Any) -> Any:
    pass

# ✅ Good - be specific
def good_func(data: dict[str, int]) -> list[str]:
    pass


# 5. USE PROTOCOL FOR DUCK TYPING
from typing import Protocol

class Closeable(Protocol):
    def close(self) -> None: ...

def cleanup(resource: Closeable) -> None:
    resource.close()
````
