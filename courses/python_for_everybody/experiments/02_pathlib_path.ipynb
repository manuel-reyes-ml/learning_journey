# Using pathlib package

## Path and Roots
Python Path object usages to get root directories

```python
from pathlib import Path

cwd = Path.cwd() # Get the current working directory
print(f"directory root: {cwd}")

file_root = cwd if (cwd / "code" / "02_data_structures" / "data").exists() else cwd.parent # Determine the file root based on the existence of a specific path
print(f"file root directory: {file_root}")

file_path = file_root / "code" / "02_data_structures" / "data"
print(f"data directory: {file_path}")
```

```python
root = Path(__file__).resolve().parents[1] # Gives Traceback in Jupyter Notebook, because the code is not being run from a .py file.
print(root)
```

```python
fname = "README.md"
p = Path(fname) # Create a Path object for the file. 'path' is now a Path object representing 'README.md' in the current directory.
print(p)
print(f"Path exists: {p.exists()}")  # Check if the path exists.

print(f"Path is a file: {p.is_file()}")  # Check if the path is a file.
```
