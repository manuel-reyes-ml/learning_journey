### 3.2 Boolean Operators

```python
# Boolean operators: and, or, not
x = 12

print(x > 5 and x < 15)   # Both must be True
print(x < 5 or x > 15)    # At least one must be True
print(not x > 5)          # Negation
```

```python
# With NumPy arrays: use logical_and, logical_or, logical_not
import numpy as np

my_array = np.array([1, 2, 3, 4, 5])

print(np.logical_and(my_array > 1, my_array < 5))  # Between 1 and 5
print(my_array[np.logical_and(my_array > 1, my_array < 5)])  # Filter
```

#### 💡 My Experiments:

```python
# Complex filtering for job search criteria
import numpy as np

# Job posting data
salaries = np.array([65, 70, 80, 55, 90, 75, 68, 85])
is_remote = np.array([True, True, False, True, False, True, True, False])
requires_experience = np.array([0, 1, 2, 0, 3, 1, 0, 2])  # Years

# Find jobs that match MY criteria:
# - Remote
# - Salary >= $68K
# - Entry level (0-1 years experience)

my_criteria = np.logical_and(
    np.logical_and(is_remote, salaries >= 68),
    requires_experience <= 1
)

print(f"Jobs matching criteria: {my_criteria}")
print(f"Number of matches: {my_criteria.sum()}")
print(f"Matching salaries: ${salaries[my_criteria]}K")
print(f"Average salary for matches: ${salaries[my_criteria].mean():.0f}K")
```

#### 💡 My Experiments:

```python
# Weekly progress evaluator
def evaluate_week(study_hours, courses_completed, problems_solved):
    """
    Evaluate my weekly performance and provide feedback
    """
    score = 0
    
    # Check study hours (goal: 25+)
    if study_hours >= 30:
        score += 3
        hours_msg = "🔥 Exceptional!"
    elif study_hours >= 25:
        score += 2
        hours_msg = "✅ Goal met!"
    elif study_hours >= 20:
        score += 1
        hours_msg = "⚠️ Close, but push harder"
    else:
        hours_msg = "❌ Need to improve"
    
    # Check courses
    if courses_completed >= 1:
        score += 2
        course_msg = "📚 Course completed!"
    else:
        course_msg = "📖 Keep studying"
    
    # Check problems
    if problems_solved >= 10:
        score += 2
        problems_msg = "💪 Great practice!"
    elif problems_solved >= 5:
        score += 1
        problems_msg = "👍 Good effort"
    else:
        problems_msg = "🎯 Need more practice"
    
    # Overall evaluation
    print(f"\n📊 WEEKLY EVALUATION")
    print(f"Study Hours ({study_hours}h): {hours_msg}")
    print(f"Courses ({courses_completed}): {course_msg}")
    print(f"Problems ({problems_solved}): {problems_msg}")
    print(f"\nTotal Score: {score}/7")
    
    if score >= 6:
        print("🌟 Outstanding week! Keep this momentum!")
    elif score >= 4:
        print("✅ Solid week. On track!")
    elif score >= 2:
        print("⚠️ Needs improvement. Refocus!")
    else:
        print("❌ Challenging week. Reset and recommit!")

# Test it
evaluate_week(study_hours=28, courses_completed=1, problems_solved=12)
evaluate_week(study_hours=22, courses_completed=0, problems_solved=4)
```

#### 💡 My Experiments:

```python
# Compound interest calculator - how long to reach goal?
initial_savings = 10000
target = 20000
annual_rate = 0.07  # 7% return

current_amount = initial_savings
years = 0

print(f"Starting with: ${current_amount:,.0f}")
print(f"Goal: ${target:,.0f}")
print(f"Annual return: {annual_rate*100}%\n")

while current_amount < target:
    years += 1
    current_amount = current_amount * (1 + annual_rate)
    print(f"Year {years}: ${current_amount:,.2f}")

print(f"\n🎯 Goal reached in {years} years!")
print(f"Final amount: ${current_amount:,.2f}")
```

### 5.2 Random Walk Simulation

```python
# Simulate a random walk
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Start at 0
position = 0
walk = [position]

# Take 100 steps
for i in range(100):
    step = np.random.randint(1, 7)  # Roll dice
    if step <= 2:
        position = max(0, position - 1)  # Can't go below 0
    else:
        position = position + 1
    walk.append(position)

# Plot
plt.plot(walk)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Random Walk')
plt.show()

print(f"Final position: {position}")
```

```python
# Simulate skill progression with realistic ups and downs
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Simulate 12 months of skill development
months = 12
skill_level = 30  # Starting proficiency
history = [skill_level]

for month in range(months):
    # Random event each month
    event = np.random.rand()
    
    if event < 0.1:  # 10% chance of setback (burnout, life events)
        change = np.random.randint(-5, -2)
        event_type = "😰 Setback"
    elif event < 0.3:  # 20% chance of plateau
        change = np.random.randint(-1, 2)
        event_type = "😐 Plateau"
    elif event < 0.7:  # 40% chance of steady progress
        change = np.random.randint(3, 7)
        event_type = "✅ Steady"
    else:  # 30% chance of breakthrough
        change = np.random.randint(8, 15)
        event_type = "🚀 Breakthrough"
    
    skill_level = min(100, max(0, skill_level + change))  # Keep between 0-100
    history.append(skill_level)
    
    print(f"Month {month+1}: {event_type:15} Change: {change:+3} → Skill: {skill_level}%")

# Visualize
plt.figure(figsize=(12, 6))
plt.plot(range(months+1), history, marker='o', linewidth=2, markersize=8)
plt.axhline(y=70, color='green', linestyle='--', alpha=0.5, label='Job-Ready (70%)')
plt.xlabel('Months', fontsize=12)
plt.ylabel('Skill Proficiency (%)', fontsize=12)
plt.title('Realistic Skill Progression Simulation\n(Including Setbacks & Breakthroughs)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(f"\n📊 Summary:")
print(f"Starting level: 30%")
print(f"Final level: {skill_level}%")
print(f"Total improvement: {skill_level - 30}%")
print(f"Average monthly gain: {(skill_level - 30)/months:.1f}%")

if skill_level >= 70:
    print(f"\n🎯 JOB-READY! Reached {skill_level}% proficiency")
else:
    print(f"\n📚 Keep learning! {70-skill_level}% to job-ready")
```

### 5.3 Monte Carlo Simulation

```python
# Run many simulations to estimate probabilities
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Run 1000 random walks
all_walks = []

for i in range(1000):
    position = 0
    walk = [position]
    
    for step in range(100):
        roll = np.random.randint(1, 7)
        if roll <= 2:
            position = max(0, position - 1)
        else:
            position = position + 1
        walk.append(position)
    
    all_walks.append(walk)

# Convert to array
all_walks = np.array(all_walks)

# Plot
plt.plot(all_walks.T, alpha=0.01, color='blue')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('1000 Random Walks')
plt.show()

# Probability of reaching 60 or higher
final_positions = all_walks[:, -1]
prob = np.sum(final_positions >= 60) / len(final_positions)
print(f"Probability of reaching 60+: {prob*100:.1f}%")
```

#### 💡 My Experiments:

```python
# Monte Carlo: What are my chances of getting a job in 5 months?
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Run 1000 simulations of my 5-month journey
num_simulations = 1000
months = 5
job_ready_threshold = 70  # Need 70% skill to be job-ready

all_journeys = []
success_count = 0

for sim in range(num_simulations):
    skill = 30  # Starting level
    journey = [skill]
    
    for month in range(months):
        # Each month has variability
        event = np.random.rand()
        
        if event < 0.1:  # 10% setback
            change = np.random.randint(-3, 0)
        elif event < 0.25:  # 15% plateau
            change = np.random.randint(0, 2)
        elif event < 0.75:  # 50% steady
            change = np.random.randint(5, 10)
        else:  # 25% breakthrough
            change = np.random.randint(10, 16)
        
        skill = min(100, max(0, skill + change))
        journey.append(skill)
    
    all_journeys.append(journey)
    
    if skill >= job_ready_threshold:
        success_count += 1

# Convert to array
all_journeys = np.array(all_journeys)

# Visualize all simulations
plt.figure(figsize=(14, 6))

# Plot 1: All journeys
plt.subplot(1, 2, 1)
plt.plot(all_journeys.T, alpha=0.02, color='blue')
plt.axhline(y=job_ready_threshold, color='green', linestyle='--', linewidth=2, label='Job-Ready (70%)')
plt.xlabel('Months')
plt.ylabel('Skill Level (%)')
plt.title(f'{num_simulations} Simulated Learning Journeys')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Distribution of final outcomes
plt.subplot(1, 2, 2)
final_skills = all_journeys[:, -1]
plt.hist(final_skills, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(x=job_ready_threshold, color='green', linestyle='--', linewidth=2, label='Job-Ready (70%)')
plt.axvline(x=final_skills.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {final_skills.mean():.1f}%')
plt.xlabel('Final Skill Level (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Final Skill Levels')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Results
success_rate = (success_count / num_simulations) * 100
print("\n" + "="*60)
print("📊 MONTE CARLO SIMULATION RESULTS")
print("="*60)
print(f"Total simulations: {num_simulations:,}")
print(f"Success threshold: {job_ready_threshold}% skill level")
print(f"Successful outcomes: {success_count:,} / {num_simulations:,}")
print(f"\n🎯 SUCCESS RATE: {success_rate:.1f}%")
print(f"\n📈 Final Skill Statistics:")
print(f"  Average: {final_skills.mean():.1f}%")
print(f"  Median: {np.median(final_skills):.1f}%")
print(f"  Std Dev: {final_skills.std():.1f}%")
print(f"  Best case: {final_skills.max():.0f}%")
print(f"  Worst case: {final_skills.min():.0f}%")
print("\n💡 Interpretation:")
if success_rate >= 80:
    print(f"  Excellent! {success_rate:.1f}% chance of success. Stay consistent!")
elif success_rate >= 60:
    print(f"  Good odds at {success_rate:.1f}%. Push through setbacks!")
elif success_rate >= 40:
    print(f"  Moderate {success_rate:.1f}% chance. Consider increasing study time.")
else:
    print(f"  Low {success_rate:.1f}% chance. Adjust strategy for better odds.")
print("="*60)
```

## 🤔 Questions for Further Learning

1. **When should I use Matplotlib vs Seaborn vs Plotly?**
   - Matplotlib: Full control, publication-quality
   - Seaborn: Statistical plots, beautiful defaults
   - Plotly: Interactive visualizations for web

2. **Why are loops "slow" in Python?**
   - Python loops have overhead per iteration
   - Vectorized operations (NumPy, pandas) use optimized C code
   - Use loops for logic, vectorization for math

3. **When should I use a dictionary vs a DataFrame?**
   - Dictionary: Small data, simple key-value mapping, config
   - DataFrame: Large datasets, need analysis/filtering, tabular data

4. **What's the difference between .loc and .iloc?**
   - `.loc[]`: Access by label (column/row names)
   - `.iloc[]`: Access by integer position (0, 1, 2...)

5. **How do I know when to use simulation?**
   - When outcomes are probabilistic/uncertain
   - When mathematical solutions are complex
   - For exploring "what-if" scenarios
