## Chapter 2: Python Lists

### Basic Lists
Lists store multiple values

```python
skills = ["Python", "SQL", "Excel", "pandas"]
print(skills)
print(skills[0])  # First item
print(skills[-1])  # Last item

# ### My Experiments:
# How many ways can I add to a list?

# Method 1: append()
skills.append("Jupyter")

# Method 2: extend()
skills.extend(["Git", "GitHub"])

# Method 3: insert()
skills.insert(1, "Data Viz")

print(f"I now have {len(skills)} skills: {skills}")
```

```python
# How many ways can I remove and item from a list?

# Method 1: remove()
skills.remove("Excel")
print(f"Method 1 after remove(): {skills}")

# Method 2: pop()
removed_skill = skills.pop(2)  # Removes "SQL" 
print(f"Method 2 after pop() - removed skill: {removed_skill}")

# Method 3: del 
del skills[0]  # Removes "Python"
print(f"Method 3 after del: {skills}")
```

```python
# Variables just point to a list in memory
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

areas_copy_1 = areas
areas_copy_2 = list(areas) #creates a new list in memory

areas_copy_1[0] = 5.0

print(f"Original areas list: {areas}")
print(f"areas_copy_1: {areas_copy_1}")
print(f"areas_copy_2: {areas_copy_2}")
```
